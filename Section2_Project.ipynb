## 나라별 홈/어웨이 승무패 및 승률 

```python
home_win, home_draw, home_lose = [], [], []
away_win, away_draw, away_lose = [], [], []

for i in team:
  #팀별 승,무,패 횟수
  h_win = df_c[(df_c["home_team"] == i)&(df_c["home_team_result"] =="Win")].count()[0]
  h_draw = df_c[(df_c["home_team"] == i)&(df_c["home_team_result"] =="Draw")].count()[0]
  h_lose = df_c[(df_c["home_team"] == i)&(df_c["home_team_result"] =="Lose")].count()[0]

  a_win = df_c[(df_c["away_team"]==i)&(df_c["home_team_result"]=="Lose")].count()[0]
  a_draw = df_c[(df_c["away_team"]==i)&(df_c["home_team_result"]=="Draw")].count()[0]
  a_lose = df_c[(df_c["away_team"]==i)&(df_c["home_team_result"]=="Win")].count()[0]

  home_win.append(h_win)
  home_draw.append(h_draw)
  home_lose.append(h_lose)

  away_win.append(a_win)
  away_draw.append(a_draw)
  away_lose.append(a_lose)
```

```python
home_total = np.array(home_win) + np.array(home_draw) + np.array(home_lose)
away_total = np.array(away_win) + np.array(away_draw) + np.array(away_lose)
```

```python
df_info = pd.DataFrame({"team" : team, "Total" :home_total + away_total,
              "Win" : np.add(home_win,away_win), "Draw": np.add(home_draw,away_draw), "Lose": np.add(home_lose,away_lose), 
              "Home_Total" : home_total,"Home_win": home_win, "Home_draw" : home_draw, "Home_lose" : home_lose,
              "Away_Total": away_total,"Away_win": away_win, "Away_draw" : away_draw, "Away_lose" : away_lose
              })
```

```python
#나라별 전체 승률
df_info.insert(5, "Total_win_rate(%)", round((df_info["Win"]/df_info["Total"])*100,3))
df_info.insert(6, "Total_draw_rate(%)", round((df_info["Draw"]/df_info["Total"])*100,3))
df_info.insert(7, "Total_lose_rate(%)", round((df_info["Lose"]/df_info["Total"])*100,3))

#나라별 홈 승률
df_info.insert(12, "Home_win_rate(%)", round((df_info["Home_win"]/df_info["Home_Total"])*100,3))
df_info.insert(13, "Home_draw_rate(%)", round((df_info["Home_draw"]/df_info["Home_Total"])*100,3))
df_info.insert(14, "Home_lose_rate(%)", round((df_info["Home_lose"]/df_info["Home_Total"])*100,3))

#나라별 어웨이 승률
df_info.insert(19, "Away_win_rate(%)", round((df_info["Away_win"]/df_info["Away_Total"])*100,3))
df_info.insert(20, "Away_draw_rate(%)", round((df_info["Away_draw"]/df_info["Away_Total"])*100,3))
df_info.insert(21, "Away_lose_rate(%)", round((df_info["Away_lose"]/df_info["Away_Total"])*100,3))
```

```python
df_info.sort_values("Total", ascending=False, inplace=True)
df_info.reset_index(drop=True, inplace=True)
df_info
```

```python
def draw_graph(df,score,title):
  fig = plt.figure(figsize=(20,10))
  fig.set_facecolor('white') 
  ax1 = fig.add_subplot()

  colors = sns.color_palette('summer', len(df["team"]))
  xtick_label_position = list(range(len(df["team"])))
  ax1.set_xticks(xtick_label_position)
  ax1.set_xticklabels(df["team"])
  ax1.bar(xtick_label_position, df[score], color=colors)

  color = 'blue'
  ax2 = ax1.twinx()
  ax2.plot(xtick_label_position, df["Total_win_rate(%)"], color=color, linestyle='--', marker='o')
  ax2.tick_params(axis='y', labelcolor=color)
  plt.title(title)
  plt.show()
```
